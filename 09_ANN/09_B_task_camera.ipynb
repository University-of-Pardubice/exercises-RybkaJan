# Objective
The objective of this exercise is to create a custom neural network that will determine what the 4 pixel camera sees.
This is the application that was introduced in the lecture.

The input will be the 4 pixel values. 
- A value of -1 represents a completely black color
- A value of +1 represents a completely white colour
- Values in between a proportional grey colour

Although the pixels are specified in a row, they are geometrically placed as follows.
- 1 2
- 3 4

The neural network should return information about whether the pixels in the image represent
- solid colour - all pixels are approximately the same colour
- vertical lines (vertical) - one line is white, the other black, in any order
- diagonal lines - there are two diagonal lines in the image, one white, the other black
- horizontal - in the picture there is one horizontal white line and one horizontal black line
